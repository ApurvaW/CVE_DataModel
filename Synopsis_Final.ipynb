```python
#import the required packages

import psycopg2
import json
import sqlalchemy
```

```python
# connecting to PostgreSQL

connection_string = 'postgresql://postgres:sushmita07@localhost:5432'
db = sqlalchemy.create_engine(connection_string)
engine = db.connect()
meta = sqlalchemy.MetaData(engine)
```

```python
# Loading the required json file in object

with open("C:/Users/Owner/Desktop/Json/nvdcve-1.1-2002.json")as load_CVE_2002:
	data_CVE_2002=json.loads(load_CVE_2002.read())
```

```python
from sqlalchemy import Column, Integer, Text
from sqlalchemy.dialects.postgresql import JSON, JSONB
```

```python
# Creating a table "CVE_2002"

sqlalchemy.Table("CVE_2002", meta,
 		   Column('CVE_data_type', Text),
 		   Column('CVE_data_format', Text),
 		   Column('CVE_data_version', Integer),
 		   Column('numberofCVEs', Integer),
 		   Column('CVE_data_timestamp', Text),
 		   Column('CVE_Items', JSON), extend_existing=True)

meta.create_all()
    
data_2002 = sqlalchemy.table(
	"CVE_2002_1", Column('CVE_data_type', Text),
		    Column('CVE_data_format', Text),
		    Column('CVE_data_version', Integer),
		    Column('numberofCVEs', Integer),
		    Column('CVE_data_timestamp', Text),
		    Column('CVE_Items', JSON))


```

```python
#inserting the required data in the fields

value = data_2002.insert().values(CVE_data_type= "CVE",
								  CVE_data_format = "MITRE",
								  CVE_data_version = 4.0,
								  numberofCVEs = 6748,
								  CVE_data_timestamp = "2020-02-25T10:02Z")
engine.execute(value)
#print(str(value))

```

```python
#using for loop to insert the CVE_Items data in the table

for i in data_CVE_2002["CVE_Items"]:
#     print(i)
#     break
    statement= data_2002.insert().values(CVE_Items= i)
    engine.execute(statement)
#     print(str(statement))
```
